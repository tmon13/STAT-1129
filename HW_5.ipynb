```r
Data_Frame1 <- data.frame(
  id = c(1,2,3,4,5),
  name = c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
  Salary = c(623.30, 515.20, 611.00, 729.00, 843.25)
)

Data_Frame1
```

```r
New_col_DF <- cbind(Data_Frame1, Department = c("Comp Sci", "Economics", "Spanish", "History", "Art"))

New_col_DF
```

```r
Data_Frame1[c(1,3,5), c(2,3)]
```

```r
salary <- Data_Frame1$Salary
salary <- salary[c(1,4,5)]

name <- Data_Frame1$name
name <- name[c(1,4,5)]

barplot(salary, names.arg=name, xlab="Name", ylab="Salary")
```

```r
salary <- Data_Frame1$Salary

high <- max(salary)
low <- min(salary)
median <- median(salary)

pie(c(high, low, median), labels=c("Michelle", "Gary", "Peter"), main="Salary Pie Chart")
```
